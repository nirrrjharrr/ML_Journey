# EDA Using Multivariate Analysis

`Multivariate Analysis` in **Exploratory Data Analysis (EDA)** involves analyzing **more than two variables at the same time** 
to understand complex relationships, interactions, and patterns within the dataset.

Unlike univariate and bivariate analysis, multivariate analysis helps reveal how multiple features jointly influence 
the target variable or each other.

---

## Objectives of Multivariate Analysis

Multivariate EDA helps to:
- Understand interactions between multiple features  
- Identify redundant or highly correlated features  
- Detect multicollinearity  
- Discover hidden patterns and clusters  
- Improve feature selection and model design

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=pd.read_csv('tips.csv')
iris=pd.read_csv('iris.csv')
flights=pd.read_csv('flights.csv')
titanic=pd.read_csv('Titanic.csv')
```

## Numerical-Numerical

### A.Scatter Plot

```python
sns.scatterplot(x='total_bill', y='tip', data=tips,hue='sex',style='smoker')
#HUE AND STYLE FUNCTION ADDED FOR MULTIVARIATE ANALYSIS
```

## Numerical-Categorical

### A.Bar Plot

```python
sns.barplot(x='Pclass', y='Fare', data=titanic,hue='Sex')
```
