# Ridge Regression (L2 Regularization)

`Ridge Regression` is a **regularized version of Linear Regression** that reduces overfitting by
penalizing large coefficient values. It adds an **L2 penalty** to the loss function, encouraging
the model to keep weights small and stable.

Ridge Regression is especially effective when features are **highly correlated** (multicollinearity).

---

## Why Ridge Regression is Needed

Ordinary Least Squares (OLS) regression minimizes:

$$
J(\theta) = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})^2
$$

This can lead to:
- Very large coefficients  
- High variance  
- Overfitting  
- Unstable predictions when features are correlated  

Ridge Regression addresses these issues by adding a penalty on coefficient magnitude.

---

## Ridge Cost Function (L2 Regularization)

The Ridge Regression objective function is:

$$
J(\theta) =
\frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})^2
+ \lambda \sum_{j=1}^{p} \theta_j^2
$$

Where:
- $(\lambda)$ = regularization strength  
- $(\theta_j)$ = model coefficients (bias term excluded)  
- $(p)$ = number of features  

The L2 penalty discourages large coefficients but does **not** force them to zero.

---

## Effect of the Regularization Parameter $(\lambda)$

| Value of $(\lambda)$ | Effect |
|---------------------|--------|
| $(\lambda = 0)$ | Same as Linear Regression |
| Small $(\lambda)$ | Mild regularization |
| Large $(\lambda)$ | Strong regularization, very small weights |

As $(\lambda)$ increases, model complexity decreases.

---

## Gradient Descent Update Rule

For Ridge Regression, parameter updates become:

$$
\theta_j := \theta_j - \alpha
\left(
\frac{\partial J}{\partial \theta_j}
+ 2\lambda \theta_j
\right)
$$

The regularization term continuously pulls coefficients toward zero,
but **never makes them exactly zero**.

---

## Key Characteristics

- Uses **L2 norm**: \( \sum \theta_j^2 \)  
- Shrinks coefficients smoothly  
- Reduces variance  
- Does **not** perform feature selection  
- Handles multicollinearity well  

---

## Importance of Feature Scaling

Feature scaling is essential for Ridge Regression because:
- The penalty depends on coefficient magnitude  
- Unscaled features can dominate regularization  

**Standardization** is strongly recommended before applying Ridge Regression.

---

## Ridge Regression vs Linear Regression

| Aspect | Linear Regression | Ridge Regression |
|------|------------------|------------------|
| Overfitting control | No | Yes |
| Penalizes coefficients | No | Yes (L2) |
| Handles multicollinearity | Poor | Good |
| Coefficients become zero | No | No |

---

## When to Use Ridge Regression

Ridge Regression is suitable when:
- The model is overfitting  
- Features are highly correlated  
- All features are expected to contribute  
- Interpretability is important but feature selection is not required  

---

## Summary

Ridge Regression improves generalization by adding an L2 penalty to the loss function,
which reduces coefficient magnitude and stabilizes the model. It is a fundamental
regularization technique for controlling overfitting in linear and polynomial regression models.

### Creating Own Ridge Regression Class (ND)

```python
class MyRidge:
    
    def __init__(self,alpha=0.1):
        
        self.alpha = alpha
        self.coef_ = None
        self.intercept_ = None
        
    def fit(self,X_train,y_train):
        
        X_train = np.insert(X_train,0,1,axis=1)
        I = np.identity(X_train.shape[1])
        I[0][0] = 0
        result = np.linalg.inv(np.dot(X_train.T,X_train) + self.alpha * I).dot(X_train.T).dot(y_train)
        self.intercept_ = result[0]
        self.coef_ = result[1:]
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

### Ridge Regression Using Gradient Descent

```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
reg = SGDRegressor(penalty='l2',max_iter=500,eta0=0.1,learning_rate='constant',alpha=0.001)
```

```python
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

```python
class MyRidgeGD:
    
    def __init__(self,epochs,learning_rate,alpha):
        
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.alpha = alpha
        self.coef_ = None
        self.intercept_ = None
        
    def fit(self,X_train,y_train):
        
        self.coef_ = np.ones(X_train.shape[1])
        self.intercept_ = 0
        thetha = np.insert(self.coef_,0,self.intercept_)
        
        X_train = np.insert(X_train,0,1,axis=1)
        
        for i in range(self.epochs):
            thetha_der = np.dot(X_train.T,X_train).dot(thetha) - np.dot(X_train.T,y_train) + self.alpha*thetha
            thetha = thetha - self.learning_rate*thetha_der
        
        self.coef_ = thetha[1:]
        self.intercept_ = thetha[0]
    
    def predict(self,X_test):
        
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
reg = MyRidgeGD(epochs=500,alpha=0.001,learning_rate=0.005)
```

```python
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```
