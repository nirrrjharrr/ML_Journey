# EDA Using Univariate Analysis

`Univariate Analysis` is a fundamental step in **Exploratory Data Analysis (EDA)** where each variable is analyzed 
**individually**, without considering relationships with other variables.  
The goal is to understand the **distribution, central tendency, spread, and anomalies** of each feature.

This step helps identify data quality issues early and guides further preprocessing and modeling decisions.

---

## Objectives of Univariate Analysis

Univariate analysis helps to:
- Understand the distribution of a feature  
- Identify missing values  
- Detect outliers  
- Understand data ranges and variability  
- Decide appropriate preprocessing techniques

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Titanic.csv')
df.head()
```

## Categorical Data

### A.Countplot

```python
sns.countplot(x='Survived',data=df)
```

### B.Piechart

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

## Numerical Data

### A.Histogram

```python
plt.hist(df['Age'],bins=10)
clean_data =  df['Age'][np.isfinite( df['Age'])]
```

### B.Displot

```python
sns.displot(df['Age'],kde=True)

```

### C.Boxplot

```python
sns.boxplot(df['Age'])
```
