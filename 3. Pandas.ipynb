# Pandas (Python Data Analysis Library)

`Pandas` is a powerful Python library used for **data manipulation and analysis**.  
It provides easy-to-use data structures that make working with structured data (tables, CSV files, Excel sheets)
simple and efficient.

In machine learning, Pandas is mainly used for **data loading, cleaning, preprocessing, and exploration**.

---

## Why Pandas is Important in Machine Learning

Pandas helps with:
- Reading datasets from files (CSV, Excel, SQL, JSON)
- Handling missing values
- Data cleaning and preprocessing
- Feature selection and transformation
- Exploratory Data Analysis (EDA)

Almost every ML project starts with Pandas before moving to NumPy or scikit-learn.

### Pandas Cheat Sheet for Data Science & ML

---

#### 1. Data Creation & Initialization
- `pd.Series(data)`: Create a Series  
- `pd.DataFrame(data)`: Create a DataFrame  
- `pd.read_csv('file.csv')`: Load CSV  
- `pd.read_excel('file.xlsx')`: Load Excel  
- `pd.DataFrame.from_dict(dict)`: Create DataFrame from dict  

---

#### 2. Data Inspection & Info
- `df.head(n)`: First n rows  
- `df.tail(n)`: Last n rows  
- `df.shape`: Dimensions  
- `df.columns`: Column names  
- `df.info()`: Data types + nulls  
- `df.describe()`: Summary stats  
- `df.dtypes`: Column types  

---

#### 3. Data Selection & Filtering
- `df['col']` / `df.col`: Select column  
- `df[['col1', 'col2']]`: Multiple columns  
- `df.iloc[0]`: Row by index  
- `df.loc[0]`: Row by label  
- `df[df['col'] > 5]`: Conditional filtering  
- `df.query('col > 5')`: Query syntax  

---

#### 4. Data Cleaning & Handling
- `df.isnull()`: Check nulls  
- `df.dropna()`: Drop nulls  
- `df.fillna(value)`: Fill nulls  
- `df.duplicated()`: Check duplicates  
- `df.drop_duplicates()`: Remove duplicates  
- `df.replace(a, b)`: Replace values  
- `df.astype(type)`: Cast data type  

---

#### 5. Aggregation & Grouping
- `df.groupby('col').mean()`: Group & aggregate  
- `df.agg(['mean', 'sum'])`: Multiple aggregations  
- `df.pivot_table(values, index, columns)`: Pivot table  
- `df.value_counts()`: Count unique values  
- `df['col'].nunique()`: Unique count  

---

#### 6. Statistical Functions
- `df.mean()`, `df.median()`, `df.std()`, `df.var()`  
- `df.corr()`: Correlation  
- `df.cov()`: Covariance  
- `df.quantile(q)`: Quantile  

---

#### 7. Combining & Merging
- `pd.concat([df1, df2], axis=0/1)`: Concatenate  
- `pd.merge(df1, df2, on='col')`: Merge  
- `df.join(df2)`: Join on index  

---

#### 8. Data Transformation
- `df.apply(func)`: Apply function  
- `df['col'].map(func)`: Map values  
- `df['col'].apply(lambda x: ...)`: Lambda apply  
- `df.rename(columns={'old': 'new'})`: Rename  
- `df.sort_values('col')`: Sort  

---

#### 9. ML Preprocessing Helpers
- `df['label'].factorize()`: Label encoding  
- `pd.get_dummies(df)`: One-hot encoding  
- `df.sample(n)`: Random sampling  
- `df.drop('col', axis=1)`: Drop column  
- `df.set_index('col')`: Set index  
- `df.reset_index()`: Reset index  

---

#### 10. Exporting Data
- `df.to_csv('file.csv', index=False)`: Save CSV  
- `df.to_excel('file.xlsx')`: Save Excel  
- `df.to_dict()`: Convert to dict
